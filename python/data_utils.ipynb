# Data Utils — Visit Date Range Filter
Reads an input CSV and keeps only rows where the visit date falls within a specified date range (inclusive on both ends).

```python
import sys
import pandas as pd
```

## Configuration
Update the paths and date range below before running.

```python
INPUT_CSV  = "path/to/input.csv"
OUTPUT_CSV = "path/to/output.csv"

DATE_COL   = "visit_date"   # name of the date column in the CSV
START_DATE = "YYYY-MM-DD"   # inclusive
END_DATE   = "YYYY-MM-DD"   # inclusive
```

## Load Input CSV

```python
df = pd.read_csv(INPUT_CSV)
print(f"Input: {len(df)} rows, {len(df.columns)} columns")
df.head()
```

## Parse & Validate Date Column

```python
if DATE_COL not in df.columns:
    raise ValueError(f"Date column '{DATE_COL}' not found. Available columns: {list(df.columns)}")

df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")

n_invalid = df[DATE_COL].isna().sum()
if n_invalid:
    print(f"Warning: {n_invalid} rows have unparseable dates and will be excluded.", file=sys.stderr)

print(f"Date range in data: {df[DATE_COL].min().date()} to {df[DATE_COL].max().date()}")
```

## Filter by Visit Date Range

```python
mask = (
    (df[DATE_COL] >= pd.Timestamp(START_DATE)) &
    (df[DATE_COL] <= pd.Timestamp(END_DATE))
)
df_out = df[mask]
print(f"Output: {len(df_out)} rows ({len(df) - len(df_out)} excluded)")
df_out.head()
```

## Save Output

```python
df_out.to_csv(OUTPUT_CSV, index=False)
print(f"Saved -> {OUTPUT_CSV}")
```
